```python
import pdfplumber
import pandas as pd
from openpyxl import Workbook
```

```python
pdf_dir = r"E:\notebooks\2.python\8.Python_PDF_ExcelReader\pdftableReader.pdf"
```

```python
# read pdf
pdf = pdfplumber.open(pdf_dir)
```

```python
# 通过pdfplumber.PDF类的metadata属性获取pdf信息
pdf.metadata
```

```python
# 总页数
len(pdf.pages)
```

```python
# 读取第一页的页宽、页高等信息

# 第一页pdfplumber.Page实例
first_page = pdf.pages[0]

# 查看页码
print('页码：', first_page.page_number)

# 查看页宽
print('页宽：', first_page.width)

# 查看页高
print('页高：', first_page.height)
```

```python
# 读取第一页文本
text = first_page.extract_text()
print(text)
```

```python
# 读取第一页表格
# 自动读取表格信息，返回列表
table = first_page.extract_tables()

# 表格转化为df
df_table = pd.DataFrame(table[1:],columns=table[0])

df_table
```
